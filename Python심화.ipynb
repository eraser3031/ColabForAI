```python
# ColabÏóêÏÑ† ÏûêÎèô ÏÑ§Ïπò.
import matplotlib.pyplot as plt
import numpy as np

x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 50]

plt.plot(x, y)  # Í∫æÏùÄÏÑ† Í∑∏ÎûòÌîÑ
plt.xlabel("X Axis")  # XÏ∂ï Î†àÏù¥Î∏î
plt.ylabel("Y Axis")  # YÏ∂ï Î†àÏù¥Î∏î
plt.title("Basic Graph")
plt.show()
```

```python
dates = ["2024-01-01", "2024-01-02", "2024-01-03", "2024-01-04"]
sales = [120, 150, 90, 200]

plt.plot(dates, sales, marker='o', linestyle='-', color='b', label="sales")
plt.xlabel("Date")
plt.ylabel("Salse (Won)")
plt.title("Changes in sales by date")
plt.legend()  # Î≤îÎ°Ä Ï∂îÍ∞Ä -o-
plt.grid(True)  # Í≤©Ïûê Ï∂îÍ∞Ä
plt.show()
```

```python
subjects = ["Math", "English", "Science", "Korean"]
scores = [85, 90, 78, 92]

plt.bar(subjects, scores, color=['red', 'blue', 'green', 'purple'])
plt.xlabel("Subjects")
plt.ylabel("Score")
plt.title("Average score by subject")
plt.ylim(0, 100)  # yÏ∂ï Î≤îÏúÑ ÏÑ§Ï†ï
plt.show()
```

```python
import pandas as pd

data = {
    "Name": ["Kim", "Lee", "Pack"],
    "Math": [85, 90, 75],
    "English": [78, 88, 80],
    "Science": [92, 95, 79]
}

df = pd.DataFrame(data)
df.set_index("Name", inplace=True)  # Ïù¥Î¶ÑÏùÑ Ïù∏Îç±Ïä§Î°ú ÏÑ§Ï†ï

df.plot(kind="bar", figsize=(10, 6))  # ÎßâÎåÄ Í∑∏ÎûòÌîÑ
plt.title("Student grade comparison")
plt.ylabel("Score")
plt.ylim(0, 100)
plt.show()
```

# 5ÏùºÏ∞® - Numpy (ÏàòÌïô Ïó∞ÏÇ∞ & ÌñâÎ†¨ Îã§Î£®Í∏∞)

üéØ Î™©Ìëú: NumpyÎ•º ÏÇ¨Ïö©Ìï¥ Î≤°ÌÑ∞ & ÌñâÎ†¨ Ïó∞ÏÇ∞ÏùÑ Î∞∞Ïö∞Í≥†, Î®∏Ïã†Îü¨ÎãùÏóêÏÑú Îç∞Ïù¥ÌÑ∞ Îã§Î£®Îäî Î≤ï ÏùµÌûàÍ∏∞

üìñ ÌïôÏäµ ÎÇ¥Ïö©

- Numpy Î∞∞Ïó¥ ÏÉùÏÑ± (np.array())
- Î∞∞Ïó¥ Ïó∞ÏÇ∞ (np.mean(), np.sum(), np.dot())
- Î≤°ÌÑ∞ & ÌñâÎ†¨ Ïó∞ÏÇ∞

üõ† Ïã§Ïäµ Î¨∏Ï†ú
- Îëê Í∞úÏùò ÌñâÎ†¨ÏùÑ Í≥±ÌïòÎäî ÏΩîÎìú ÏûëÏÑ±
- 1Ï∞®Ïõê Î∞∞Ïó¥Ïùò ÌèâÍ∑†Í≥º Ìï© Íµ¨ÌïòÍ∏∞

üëâ Numpy(ÎÑòÌååÏù¥)Îäî ÏàòÌïô Ïó∞ÏÇ∞ÏùÑ Îπ†Î•¥Í≤å Ï≤òÎ¶¨ÌïòÎäî ÎùºÏù¥Î∏åÎü¨Î¶¨

üëâ PandasÍ∞Ä Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ(Ïó¥, Ìñâ 2Ï∞®Ïõê Îç∞Ïù¥ÌÑ∞Íµ¨Ï°∞)ÏùÑ Îã§Î£®Îäî ÎùºÏù¥Î∏åÎü¨Î¶¨ÎùºÎ©¥, NumpyÎäî **ÏàòÌïô Í≥ÑÏÇ∞(Î≤°ÌÑ∞, ÌñâÎ†¨, ÌÜµÍ≥Ñ)**ÏùÑ Îã§Î£®Îäî ÎèÑÍµ¨!

üìå NumpyÏùò ÌäπÏßï

- ÏùºÎ∞ò Î¶¨Ïä§Ìä∏Î≥¥Îã§ Ïó∞ÏÇ∞ ÏÜçÎèÑÍ∞Ä Ìõ®Ïî¨ Îπ†Î¶Ñ
- Î≤°ÌÑ∞, ÌñâÎ†¨ Ïó∞ÏÇ∞ ÏßÄÏõê (NumPy Î∞∞Ïó¥)
- Í≥ºÌïô Í≥ÑÏÇ∞, Î®∏Ïã†Îü¨Îãù, Îî•Îü¨ÎãùÏóêÏÑú ÌïÑÏàò


```python
import numpy as np  # Colab Í∏∞Î≥∏ ÏÑ§Ïπò
```

```python
# Î¶¨Ïä§Ìä∏Î•º Numpy Î∞∞Ïó¥Î°ú Î≥ÄÌôò
arr = np.array([1, 2, 3, 4, 5])

print(arr)
print(type(arr))  # Ï∂úÎ†•: <class 'numpy.ndarray'>
```

```python
matrix = np.array([[1, 2, 3], [4, 5, 6]])  # 2x3 ÌñâÎ†¨
print(matrix)
```

```python
# for Î¨∏ ÏóÜÏù¥ Îπ†Î•¥Í≤å Ïó∞ÏÇ∞, ÏùºÎ∞ò Î¶¨Ïä§Ìä∏Î≥¥Îã§ Ïó∞ÏÇ∞ÏÜçÎèÑ Ìõ®Ïî¨ Îπ†Î¶Ñ

# Ìï©
arr = np.array([1, 2, 3, 4, 5])
print(np.sum(arr))  # Ï∂úÎ†•: 15
# ÌèâÍ∑†
print(np.mean(arr))  # Ï∂úÎ†•: 3.0
# ÏµúÎåÄ ÏµúÏÜå
print(np.max(arr))  # Ï∂úÎ†•: 5
print(np.min(arr))  # Ï∂úÎ†•: 1
# ÏöîÏÜåÎÅºÎ¶¨ Ïó∞ÏÇ∞
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])
print(arr1 + arr2)  # Ï∂úÎ†•: [5 7 9]
print(arr1 * arr2)  # Ï∂úÎ†•: [4 10 18] (Í∞Å ÏõêÏÜåÎÅºÎ¶¨ Í≥±ÏÖà)
```

```python
# Î≤°ÌÑ∞ Î∞è ÌñâÎ†¨ Ïó∞ÏÇ∞

# ÌñâÎ†¨Í≥±
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

C = np.dot(A, B)  # ÌñâÎ†¨ Í≥±
print(C)
print(f"ÏõêÏÜåÎÅºÎ¶¨ Í≥±:\n {A * B}")  # ÏõêÏÜåÎ≥Ñ Í≥±ÏÖà !Ï£ºÏùò
```

```python
# Ï†ÑÏπòÌñâÎ†¨
A = np.array([[1, 2], [3, 4]])
print(np.transpose(A))  # ÌñâÎ†¨ Ï†ÑÏπò (Ìñâ ‚Üî Ïó¥ Î∞îÍæ∏Í∏∞) Ï∂ï Í∏∞Ï§ÄÏúºÎ°ú Îí§ÏßëÎäîÎã® ÏÉùÍ∞Å

# Ïó≠ÌñâÎ†¨
A = np.array([[4, 7], [2, 6]])
A_inv = np.linalg.inv(A)  # Ïó≠ÌñâÎ†¨ Í≥ÑÏÇ∞ (Í∏∞Ï°¥ ÌñâÎ†¨ Í≥±ÌñàÏùÑ Îïå, Îã®ÏúÑÌñâÎ†¨ ÎêòÎäî Ïï†)
print(A_inv)
print(np.dot(A, A_inv))

# ÌñâÎ†¨Ïãù
det_A = np.linalg.det(A)  # ÌñâÎ†¨Ïãù Í≥ÑÏÇ∞
print(det_A)

```

 PandasÎäî ÌååÏù¥Ïç¨ÏóêÏÑú Îç∞Ïù¥ÌÑ∞Î•º ÏâΩÍ≤å Îã§Î£∞ Ïàò ÏûàÎèÑÎ°ù ÎèÑÏôÄÏ£ºÎäî ÎùºÏù¥Î∏åÎü¨Î¶¨

üëâ ÏóëÏÖÄÏ≤òÎüº Ìëú(ÌÖåÏù¥Î∏î) ÌòïÏãùÏùò Îç∞Ïù¥ÌÑ∞Î•º Ï≤òÎ¶¨Ìï† Ïàò ÏûàÎã§!

üëâ Î®∏Ïã†Îü¨ÎãùÍ≥º Îç∞Ïù¥ÌÑ∞ Î∂ÑÏÑùÏóêÏÑú Î∞òÎìúÏãú ÏïåÏïÑÏïº ÌïòÎäî ÌïÑÏàò ÎèÑÍµ¨.

```python
import pandas as pd  # Pandas Î∂àÎü¨Ïò§Í∏∞. Ïó≠ÏãúÎÇò Í∏∞Î≥∏ ÏÑ§ÏπòÎê®.
```

```python
# PandasÏùò ÌïµÏã¨ Í∞ùÏ≤¥Îäî DataFrame(2Ï∞®Ïõê Ìëú ÌòïÌÉú Îç∞Ïù¥ÌÑ∞)ÏôÄ Series(1Ï∞®Ïõê Îç∞Ïù¥ÌÑ∞)

data = {
    "Ïù¥Î¶Ñ": ["ÍπÄÏ≤†Ïàò", "Ïù¥ÏòÅÌù¨", "Î∞ïÎØºÏàò"],
    "ÏàòÌïô": [85, 90, 75],
    "ÏòÅÏñ¥": [78, 88, 80],
    "Í≥ºÌïô": [92, 95, 79]
}

df = pd.DataFrame(data)  # Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ ÏÉùÏÑ±
print(df)
```

```python
df.set_index("Ïù¥Î¶Ñ", inplace=True) # ÏúÑÏóêÎäî Ïà´ÏûêÍ∞Ä Ïù∏Îç±Ïä§,
# inplace ÌÇ§Î©¥ ÏõêÎ≥∏ ÏßÅÏ†ë Î≥ÄÍ≤Ω. df ÏûêÏ≤¥ Î∞îÎÄú. ÎÅÑÎ©¥ df2 = df.set_index("Ïù¥Î¶Ñ") Ìï¥ÏÑú Îî∞Î°ú Ï†ÄÏû•Ìï¥ÏïºÌï®.
print(df)
```

```python
# CSV ÌååÏùºÎ°ú Ï†ÄÏû• (Ïù∏Îç±Ïä§ Ï†úÏô∏)
df.to_csv("students.csv", index=True)

# CSV ÌååÏùº Î∂àÎü¨Ïò§Í∏∞
df = pd.read_csv("students.csv")  # CSV ÌååÏùº ÏùΩÍ∏∞
print(df.head())  # ÏÉÅÏúÑ 5Í∞ú Ìñâ Ï∂úÎ†•
```

```python
print(df.describe())  # ÌèâÍ∑†, ÌëúÏ§ÄÌé∏Ï∞®, ÏµúÎåìÍ∞í, ÏµúÏÜüÍ∞í Îì± Ï∂úÎ†•
print("\n")
print(f"ÌèâÍ∑†Í∞í = {df['ÏàòÌïô'].mean()}")  # ÏàòÌïô Ï†êÏàòÏùò ÌèâÍ∑†Í∞í
print("\n")
print(df[df["ÏàòÌïô"] >= 80])  # ÏàòÌïô Ï†êÏàòÍ∞Ä 80Ï†ê Ïù¥ÏÉÅÏù∏ ÌïôÏÉùÎßå Ï∂úÎ†•
print("\n")
print(df.groupby("Ïù¥Î¶Ñ")["ÏàòÌïô"].mean())  #Ïù¥Î¶ÑÎ≥Ñ ÏàòÌïô Ï†êÏàò ÌèâÍ∑†(ÏùòÎØ∏ÏóÜÏßÄÎßå ÎßåÏïΩ ÌïôÎÖÑÎ≥Ñ Ïù¥Îû¨ÏúºÎ©¥ ÏùòÎØ∏ ÏûàÍ≤†ÏßÄ.)
print("\n")
df.fillna(0, inplace=True)  # NaN(Í≤∞Ï∏°Ïπò)ÏùÑ ÌäπÏ†ïÍ∞íÏúºÎ°ú Ï±ÑÏö∞Í∏∞
df["Ï¥ùÏ†ê"] = df["ÏàòÌïô"] + df["ÏòÅÏñ¥"] + df["Í≥ºÌïô"]  # Ï¥ùÏ†ê Ïó¥ Ï∂îÍ∞Ä
print(df)
df.drop(columns=["Ï¥ùÏ†ê"], inplace=True)  # 'Ï¥ùÏ†ê' Ïó¥ ÏÇ≠Ï†ú
print("\n")
print(df)
```
